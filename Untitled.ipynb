```python
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import numpy as np

def moving_average(x, N):
#     print(len(x))
#     return x
    return np.convolve(x, np.ones(N, ), mode='valid') / N

```

```python
N_elems = 10000
SLIDING_N = 40
N_elems_2 = len(moving_average(np.ones(N_elems), N_elems//SLIDING_N))
# print(N_elems)
N_exps = 10
total = N_elems_2 * N_exps
arr_maddpg = np.zeros((total, 3))
arr_td3 = np.zeros((total, 3))
arr_td3_clip05 = np.zeros((total, 3))
arr_td3_clip02 = np.zeros((total, 3))
arr_ddpg = np.zeros((total, 3))
arr_ddpg_all_obs = np.zeros((total, 3))
td3_longer_rews = np.zeros((total, 3))

def get_results_arr(base_name, idx):
#     arr = 
    idxes = (np.zeros(total) + idx).reshape(-1, 1)
#     a = moving_average(np.recfromcsv(base_name + str(i) + '.csv').astype('float32'), N_elems//10)
#     print(len(a))
    arr = [np.vstack([np.arange(N_elems_2), moving_average(np.recfromcsv(base_name + \
                        str(i) + '.csv').astype('float32'), N_elems//SLIDING_N)]).T for i in range(10)]
    return np.hstack([idxes, np.vstack(arr)])
    
    
# for i in range(10):
#     "td3_better_smooting-10k7.csv"
#     "td3_better_smooting-target_clip02-10k2.csv"
# arr_td3 = np.recfromcsv('ep_rewards/Td3-10k' + str(i) + '.csv')
# arr_td3_clip05[:, i] = np.recfromcsv("ep_rewards/td3_better_smooting-10k" + str(i) + '.csv')
# arr_td3_clip02[:, i] = np.recfromcsv('ep_rewards/td3_better_smooting-target_clip02-10k' + str(i) + '.csv')
# arr_ddpg[:, i] = np.recfromcsv('ep_rewards/DDPG' + str(i) + '.csv')
# arr_ddpg_all_obs[:, i] = np.recfromcsv('ep_rewards/DDPG_all_obs' + str(i) + '.csv')
# td3_longer_rews[:, i] = np.recfromcsv('ep_rewards/TD3-longer_discrete' + str(i) + '.csv')
# var_maddpg = np.var(arr_maddpg, 1)
# mean_maddpg = np.mean(arr_maddpg, 1)
# var_td3 = np.var(arr_td3, 1)
# mean_td3 = np.mean(arr_td3, 1)
arr_maddpg = get_results_arr('ep_rewards/MADDPG_rewards10k', 0)
arr_ddpg_all_obs = get_results_arr('ep_rewards/DDPG_all_obs', 1)
td3_longer_rews = get_results_arr('ep_rewards/TD3-longer_discrete', 2)

# arr_td3 = get_results_arr('ep_rewards/td3_better_smooting-10k', 1)
# arr_td3_clip05 = get_results_arr('ep_rewards/td3_better_smooting-10k', 2)
# arr_td3_clip02 = get_results_arr('ep_rewards/td3_better_smooting-target_clip02-10k', 3)
arr_ddpg = get_results_arr('ep_rewards/DDPG', 3)


all_data = np.vstack([
    arr_maddpg, 
#     arr_td3, 
#     arr_td3_clip05, 
#     arr_td3_clip02, 
    arr_ddpg_all_obs, 
    td3_longer_rews,
    arr_ddpg
])
```

```python

data = pd.DataFrame(all_data, columns=["Algorithm", "Episode", "Reward"])
data = data.astype({"Algorithm": str, "Episode": int, "Reward": float})
print(data)
data.loc[data['Algorithm'] == "0.0", 'Algorithm'] = "MADDPG"
data.loc[data['Algorithm'] == "1.0", 'Algorithm'] = "DDPG all obs"
data.loc[data['Algorithm'] == "2.0", 'Algorithm'] = "TD3-frequent updates"
data.loc[data['Algorithm'] == "3.0", 'Algorithm'] = "DDPG"

# print(data)
# sns.lineplot(data, hue=)
# sns.set_context("talk")
plt.figure(figsize=(20,20))

sns.set(font_scale=2.3)  # crazy big
font = {'family' : 'normal',
        'weight' : 'bold'
       }
plt.tight_layout()
matplotlib.rc('font', **font)
plt.ylim(-100, 0)
plt.legend() 
# import torch.nn as nn
sns.lineplot(data=data, hue="Algorithm", x="Episode", y="Reward", 
             palette=['red', 'gray', 'blue', 'green'])
plt.savefig("Better_plot.png")
plt.show()
```

```python
sns.set_context("talk")

window = int(len(arr_maddpg)/30)
def get_plot_arrs(arr, window):
    return moving_average(np.sqrt(np.var(arr, 1)), window), moving_average(np.mean(arr, 1), window)

plt.figure(figsize=(20,20))
var_maddpg, mean_maddpg = get_plot_arrs(arr_maddpg, window)
x = np.arange(len(mean_maddpg))
plt.plot(x, mean_maddpg, label="MADDPG")
# plt.fill_between(x, mean_maddpg - var_maddpg, mean_maddpg + var_maddpg, alpha=0.2)


var_td3, mean_td3 = get_plot_arrs(arr_td3, window)
x = np.arange(len(mean_td3))
plt.plot(x, mean_td3, label="TD3 discrete random a'")
# plt.fill_between(x, mean_td3 - var_td3, mean_td3 + var_td3, alpha=0.2)


var_td3_5, mean_td3_5 = get_plot_arrs(arr_td3_clip05, window)
x = np.arange(len(mean_td3_5))
plt.plot(x, mean_td3_5, label="TD3-0.5 continuous clip")
# plt.fill_between(x, mean_td3_5 - var_td3_5, mean_td3_5 + var_td3_5, alpha=0.2)

var_td3_2, mean_td3_2 = get_plot_arrs(arr_td3_clip02, window)
x = np.arange(len(mean_td3_2))
plt.plot(x, mean_td3_2, label="TD3-0.2 continuous clip")
# plt.fill_between(x, mean_td3_2 - var_td3_2, mean_td3 + var_td3_2, alpha=0.2)

var_ddpg, mean_ddpg = get_plot_arrs(arr_ddpg, window)
x = np.arange(len(mean_ddpg))
plt.plot(x, mean_ddpg, label="DDPG")
# plt.fill_between(x, mean_ddpg - var_ddpg, mean_ddpg + var_ddpg, alpha=0.2)

var_arr_ddpg_all_obs, mean_arr_ddpg_all_obs = get_plot_arrs(arr_ddpg_all_obs, window)
x = np.arange(len(mean_arr_ddpg_all_obs))
ax = plt.plot(x, mean_arr_ddpg_all_obs, label="DDPG all observations")
# plt.fill_between(x, mean_arr_ddpg_all_obs - var_arr_ddpg_all_obs, mean_arr_ddpg_all_obs + var_arr_ddpg_all_obs, alpha=0.2)

var_td3_l, mean_td3_l = get_plot_arrs(td3_longer_rews, window)
x = np.arange(len(mean_td3_l))
ax = plt.plot(x, mean_td3_l, label="TD3-more frequent train")

plt.xlabel("Episodes", fontdict={"fontsize": 'x-large'})
plt.ylabel("Episode rewards", fontdict={"fontsize": 'x-large'})
plt.title("Comparison of RL algorithms on cooperative communication", fontdict={"fontsize": 'x-large'}, pad=40) 
# plt.set_title(pad=20)
sns.set(font_scale=2.3)  # crazy big
font = {'family' : 'normal',
        'weight' : 'bold'
       }
plt.tight_layout()
matplotlib.rc('font', **font)
plt.ylim(-150, 0)
plt.legend()
plt.savefig("Comparison_plot.png")
plt.show()
```
